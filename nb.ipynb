```python
import os


class BaseChain:
    def __init__(self, template):
        self.llm = ChatGroq(
    temperature=0,
    model= "llama-3.1-70b-versatile", #"llama3-70b-8192",
    api_key="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",
    verbose= True,
    max_retries=2,
)
        self.prompt = PromptTemplate(input_variables=["input"], template=template)
        self.chain = LLMChain(llm=self.llm, prompt=self.prompt)

    def run(self, input_text):
        return self.chain.run(input_text)
```

```python
class TaskParserChain(BaseChain):
    def __init__(self):
        template = """
        Extract and list all tasks from the following project description:

        {input}

        Tasks:
        """
        super().__init__(template)
```

```python
class UserStoryExtractorChain(BaseChain):
    def __init__(self):
        template = """
        Given the following project requirement, extract and format user stories:

        {input}

        Please format each user story as:
        "As a [type of user], I want [an action] so that [a benefit/a value]."

        User Stories:
        """
        super().__init__(template)
```

```python
class BacklogPrioritizerChain(BaseChain):
    def __init__(self):
        template = """
        Prioritize the following backlog items based on business value, effort, and dependencies:

        {input}

        Provide a numbered list of prioritized items with brief explanations:
        """
        super().__init__(template)
```

```python
class SprintPlannerChain(BaseChain):
    def __init__(self):
        template = """
        Given the following prioritized backlog and team capacity, create a sprint plan:

        Backlog:
        {input}

        Team Capacity: 100 story points

        Provide a sprint plan with selected items and their estimated story points:
        """
        super().__init__(template)
```

```python
class AgileProjectManager:
    def __init__(self):
        self.task_parser = TaskParserChain()
        self.user_story_extractor = UserStoryExtractorChain()
        self.backlog_prioritizer = BacklogPrioritizerChain()
        self.sprint_planner = SprintPlannerChain()

    def process_project(self, project_description):
        tasks = self.task_parser.run(project_description)
        user_stories = self.user_story_extractor.run(project_description)
        backlog = f"Tasks:\n{tasks}\n\nUser Stories:\n{user_stories}"
        prioritized_backlog = self.backlog_prioritizer.run(backlog)
        sprint_plan = self.sprint_planner.run(prioritized_backlog)
        return {
            "tasks": tasks,
            "user_stories": user_stories,
            "prioritized_backlog": prioritized_backlog,
            "sprint_plan": sprint_plan
        }
```

```python
import os
from typing import Dict, Any
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage


class AgileProjectManagerModel:
    def __init__(self, model_name: str = "llama-3.1-70b-versatile", api_key: str = None):
        self.llm = ChatGroq(
            temperature=0,
            model=model_name,
            api_key=api_key or os.getenv("GROQ_API_KEY"),
            verbose=True,
            max_retries=2,
        )
        self.templates = self._initialize_templates()

    def _initialize_templates(self) -> Dict[str, str]:
        return {
            "task_parser": """
            Extract and list all tasks from the following project description:

            {input}

            Tasks:
            """,
            "user_story_extractor": """
            Given the following project requirement, extract and format user stories:

            {input}

            Please format each user story as:
            "As a [type of user], I want [an action] so that [a benefit/a value]."

            User Stories:
            """,
            "backlog_prioritizer": """
            Prioritize the following backlog items based on business value, effort, and dependencies:

            {input}

            Provide a numbered list of prioritized items with brief explanations:
            """,
            "sprint_planner": """
            Given the following prioritized backlog and team capacity, create a sprint plan:

            Backlog:
            {input}

            Team Capacity: 100 story points

            Provide a sprint plan with selected items and their estimated story points:
            """
        }


    def _run_chain(self, template_key: str, input_text: str) -> str:
        prompt = self.templates[template_key].format(input=input_text)
        
        # Create a proper message object
        message = HumanMessage(content=prompt)
        
        try:
            # Generate the response
            response = self.llm.generate([[message]])
            
            # Extract the generated content
            if response.generations and response.generations[0]:
                return response.generations[0][0].text
            else:
                raise ValueError("No content generated by the model")
        
        except AttributeError:
            # If 'generate' method is not available, try 'predict' method
            try:
                return self.llm.predict(prompt)
            except AttributeError:
                raise AttributeError("The language model does not have a supported method for generating responses.")
        
        except Exception as e:
            print(f"An error occurred while generating response: {str(e)}")
            raise

    def process_project(self, project_description: str) -> Dict[str, str]:
        tasks = self._run_chain("task_parser", project_description)
        user_stories = self._run_chain("user_story_extractor", project_description)
        backlog = f"Tasks:\n{tasks}\n\nUser Stories:\n{user_stories}"
        prioritized_backlog = self._run_chain("backlog_prioritizer", backlog)
        sprint_plan = self._run_chain("sprint_planner", prioritized_backlog)
        
        return {
            "tasks": tasks,
            "user_stories": user_stories,
            "prioritized_backlog": prioritized_backlog,
            "sprint_plan": sprint_plan
        }

    def save_model(self, filepath: str):
        # In a real scenario, you'd save model weights, hyperparameters, etc.
        # For this example, we'll just save the templates
        import json
        with open(filepath, 'w') as f:
            json.dump(self.templates, f)

    @classmethod
    def load_model(cls, filepath: str, model_name: str = "llama-3.1-70b-versatile", api_key: str = None):
        model = cls(model_name, api_key)
        import json
        with open(filepath, 'r') as f:
            model.templates = json.load(f)
        return model

def main():
    api_key = "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"
    model = AgileProjectManagerModel(api_key=api_key)
    
    project_description = """
    Project Title: Wellness Buddy: Your Personal Health & Wellness Companion
    Overview:
    Wellness Buddy is a mobile app designed to help users manage their physical and mental well-being. The app features daily wellness tracking, personalized health tips, guided exercises, and a chatbot, "Buddy," that acts as a virtual wellness assistant. By integrating AI-driven chatbot technology, Wellness Buddy provides users with real-time advice and support for their health goals.

    Features:

    Wellness Tracking: Users can log their daily activities such as water intake, meals, exercise, sleep patterns, and mood. The app provides insights and trends over time to help users maintain healthy habits.
    Personalized Health Tips: Based on user data and preferences, the app suggests personalized health tips, meal plans, and workout routines.
    Guided Exercises & Meditation: The app includes a variety of guided exercises, stretching routines, and mindfulness meditations to enhance physical and mental wellness.
    Buddy Chatbot: The chatbot feature serves as a 24/7 virtual assistant, allowing users to ask questions about fitness, nutrition, mental health, and more. Buddy can provide personalized advice, suggest exercises, recommend healthy recipes, and offer mindfulness techniques.
    """

    result = model.process_project(project_description)

    print("Tasks:")
    print(result["tasks"])
    print("\nUser Stories:")
    print(result["user_stories"])
    print("\nPrioritized Backlog:")
    print(result["prioritized_backlog"])
    print("\nSprint Plan:")
    print(result["sprint_plan"])

    # Save and load model demonstration
    model.save_model("agile_project_manager_model.json")
    loaded_model = AgileProjectManagerModel.load_model("agile_project_manager_model.json", api_key=api_key)
    loaded_result = loaded_model.process_project(project_description)
    print("\nLoaded Model Result (should be similar to the original):")
    print(loaded_result)

if __name__ == "__main__":
    main()
```
